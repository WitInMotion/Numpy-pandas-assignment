## Part A: NumPy Array Operations  
### Question 1: Revenue Simulation

```python
# Set random seed
np.random.seed(7)

# Generate daily sales quantities for 30 days
product_a = np.random.normal(50, 10, 30)
product_b = np.random.normal(30, 5, 30)
product_c = np.random.normal(70, 15, 30)
product_d = np.random.normal(40, 8, 30)

# Ensure no negative values
product_a = np.clip(product_a, 0, None)
product_b = np.clip(product_b, 0, None)
product_c = np.clip(product_c, 0, None)
product_d = np.clip(product_d, 0, None)

# Combine into 30x4 array
daily_sales = np.column_stack(
    (product_a, product_b, product_c, product_d)
)

# Display results
print("Shape of daily_sales:", daily_sales.shape)
print("First 5 days of sales:\n", daily_sales[:5])
```

### Question 2: Financial Analysis

```python
# Total sales per product
total_sales_per_product = daily_sales.sum(axis=0)

# Average daily sales per product
average_sales_per_product = daily_sales.mean(axis=0)

# Total sales per day
total_sales_per_day = daily_sales.sum(axis=1)

# Day with maximum total sales
max_sales_day = np.argmax(total_sales_per_day) + 1  # +1 for human-friendly day

# Number of days above average total sales
average_total_sales = total_sales_per_day.mean()
days_above_average = np.sum(total_sales_per_day > average_total_sales)

# Standard deviation per product
std_sales_per_product = daily_sales.std(axis=0)

# Print results
print("Total sales per product:", total_sales_per_product)
print("Average daily sales per product:", average_sales_per_product)
print("Day with highest total sales:", max_sales_day)
print("Days with sales above average:", days_above_average)
print("Standard deviation per product:", std_sales_per_product)
```

### Question 3: Price and Revenue Calculation

```python
prices = np.array([200, 350, 150, 280])

# Daily revenue without discount
daily_revenue = daily_sales * prices
total_revenue_no_discount = daily_revenue.sum()

# Copy revenue for discount calculations
discounted_revenue = daily_revenue.copy()

# Days 1–5: 15% discount on all products
discounted_revenue[0:5, :] *= 0.85

# Days 15–20: 10% discount on Product A & B
discounted_revenue[14:20, 0:2] *= 0.90

# Total revenue with discounts
total_revenue_with_discount = discounted_revenue.sum()

# Discount given
discount_given = total_revenue_no_discount - total_revenue_with_discount

print("Total revenue (no discount):", total_revenue_no_discount)
print("Total revenue (with discount):", total_revenue_with_discount)
print("Total discount given:", discount_given)
```

```python
print("Shape:", df.shape)
print("\nInfo:")
df.info()

print("\nFirst 10 rows:")
df.head(10)

print("\nStatistical summary:")
df.describe()

print("\nMissing values:")
df.isna().sum()
```

This dataset contains information on Nigerian market products, including prices, locations, and categories.
Each row represents a product entry in a specific market or state.
The data is useful for analyzing pricing patterns and regional differences.

```python
# Select columns
df[['product', 'price', 'state']].head(10)

# Convert 'price' to numeric
df['price'] = df['price'].astype(str).str.replace('₦', '', regex=False).str.replace(',', '', regex=False)
df['price'] = pd.to_numeric(df['price'], errors='coerce')
# Fill NaN values with 0 and convert to integer
df['price'] = df['price'].fillna(0).astype(int)
# Ensure no negative prices
df['price'] = df['price'].apply(lambda x: max(0, x))

# Numeric filter
price_filter = df[df['price'] > 500]
print("Price > 500 shape:", price_filter.shape)

# Categorical filter
lagos_filter = df[df['state'] == 'Lagos']
print("Lagos shape:", lagos_filter.shape)

# Combined filter
combined_filter = df[(df['price'] > 500) & (df['state'] == 'Lagos')]
print("Combined shape:", combined_filter.shape)
```

```python
# New column
# Add a placeholder 'quantity' column as it's missing from the DataFrame
df['quantity'] = 1  # Assuming a default quantity of 1 for now
df['total_value'] = df['price'] * df['quantity']

# Categorize price
df['price_level'] = pd.cut(
    df['price'],
    bins=[0, 200, 500, np.inf],
    labels=['Low', 'Medium', 'High']
)

# Sort
df.sort_values(['state', 'price'], ascending=[True, False])

# Cumulative quantity
df['cumulative_quantity'] = df['quantity'].cumsum()

# Handle missing values using the recommended ffill method
df.ffill(inplace=True)

df.head()
```

Question 8a: Which product has the highest average price across all states?

```python
avg_price_by_product = df.groupby('product')['price'].mean().sort_values(ascending=False)
avg_price_by_product
```

```python
avg_price_by_product.plot(kind='bar', title='Average Price by Product')
```

Interpretation
The product with the highest average price is the most premium item in the market.
This insight helps sellers understand which products contribute more per unit sale and may require different pricing or stocking strategies.

## Question 8b: Which state has the highest total market value based on price and quantity?

```python
state_market_value = df.groupby('state')['total_value'].sum().sort_values(ascending=False)
state_market_value
```

```python
state_market_value.plot(kind='bar', title='Total Market Value by State')
```

### Interpretation
The state with the highest total market value represents the strongest market in terms of sales volume and pricing.
This information is useful for prioritizing expansion, logistics planning, and targeted marketing efforts.

## Question 8c: How are products distributed across price categories (Low, Medium, High)?

```python
price_distribution = df['price_level'].value_counts()
price_distribution
```

```python
price_distribution.plot(kind='pie', autopct='%1.1f%%', title='Price Level Distribution')
```

### Interpretation
Most products fall into the high price category, indicating an imbalanced market focused on affordability.
This suggests that extreme low or medium pricing is less common, which can guide pricing decisions and product positioning.
